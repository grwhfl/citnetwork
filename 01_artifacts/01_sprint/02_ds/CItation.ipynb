```python
import pandas as pd
```

```python
with open('DBLP-citation-Feb21.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()
```

```python
#* --- paperTitle
#@ --- Authors
#year ---- Year
#conf --- publication venue
#citation --- citation number (both -1 and 0 means none)
#index ---- index id of this paper
#arnetid ---- pid in arnetminer database
#% ---- the id of references of this paper (there are multiple lines, with each indicating a reference)
#! --- Abstract
```

```python
lines[:10]
```

```python
l = []
new = []
for item in lines[1:]:
    if item == "\n":
        l.append(new)
        new = []
    else:
        new.append(item)
```

```python
l[:2]
```

```python
len(l)
```

```python
def list_to_dict(lst):
    
    dct = {'papertitle': '',
     'authors': '',
     'year': '',
     'conf': '',
     'citation': '',
     'index': '',
     'arnetid': '',
     'id_of_references': [],
     'abstract': ''}
    
    for s in lst:
        if s.startswith("#*"):
            dct['papertitle'] = s[2:].replace('\n', '')
        if s.startswith("#@"):
            dct['authors'] = s[2:].replace('\n', '')
        if s.startswith("#year"):
            dct['year'] = s[5:].replace('\n', '')
        if s.startswith("#conf"):
            dct['conf'] = s[5:].replace('\n', '')
        if s.startswith("#citation"):
            dct['citation'] = s[9:].replace('\n', '')
        if s.startswith("#index"):
            dct['index'] = s[6:].replace('\n', '')
        if s.startswith("#arnetid"):
            dct['arnetid'] = s[8:].replace('\n', '')
        if s.startswith("#%"):
            dct['id_of_references'].append(s[2:].replace('\n', ''))
        if s.startswith("#!"):
            dct['abstract'] = s[2:].replace('\n', '')

    return dct
```

```python
from tqdm import tqdm
```

```python
data = [list_to_dict(item) for item in tqdm(l)]
```

```python
data
```

```python
import json

with open("data.json", "w") as write_file:
    json.dump(data, write_file, indent=4)
```

```python
df = pd.DataFrame(data)
```

```python
del lines
del l
del data
```
